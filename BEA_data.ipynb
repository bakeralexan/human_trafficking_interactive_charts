```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_file = "Resources/BEA_data.csv"
BEA_df = pd.read_csv(csv_file, encoding='latin-1')
BEA_df.head()
```

```python
reduced_BEA_df = BEA_df[["GeoFIPS", "GeoName", "Region", "TableName", "LineCode", "IndustryClassification", "Description", "Unit", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020" ]]
reduced_BEA_df
```

```python
reduced_BEA_df.to_csv('BEA_2013_2020.csv', encoding='utf-8', index=False)
```

```python
reduced_BEA_df.dropna(how = 'any', inplace = True)
```

```python
#splitting County And State
reduced_BEA_df['County'], reduced_BEA_df['State'] = reduced_BEA_df ['GeoName'].str.split(', ', 1).str
reduced_BEA_df
```

```python
income_per_capita_df = reduced_BEA_df.loc[reduced_BEA_df['LineCode'] == 3.0]
income_per_capita_df
```

```python
reduced_income_per_capita_df = income_per_capita_df.loc[(income_per_capita_df['2013']!= "(NA)") & (income_per_capita_df['2014']!= "(NA)")
                                                       & (income_per_capita_df['2015']!= "(NA)") & (income_per_capita_df['2016']!= "(NA)")
                                                       & (income_per_capita_df['2017']!= "(NA)") & (income_per_capita_df['2018']!= "(NA)") & (income_per_capita_df['2019']!= "(NA)") & (income_per_capita_df['2020']!= "(NA)")]

reduced_income_per_capita_df.dropna(how = 'any', inplace = True)
reduced_income_per_capita_df
```

```python
reduced_income_per_capita_df[['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']] = reduced_income_per_capita_df[['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']].apply(pd.to_numeric)
```

```python
list_of_states = reduced_income_per_capita_df['State'].unique()
list_of_states
```

```python
# removing rows from the df with state name more than two characters

reduced_income_per_capita_df['state_length'] = reduced_income_per_capita_df.State.str.len()
reduced_income_per_capita_df_2 = reduced_income_per_capita_df.loc[reduced_income_per_capita_df['state_length']<3]
reduced_income_per_capita_df_2
```

```python
reduced_income_per_capita_df_2.to_csv('BEA_2013_2020_clean.csv', encoding='utf-8', index=False)
```

```python
#median income per capita grouped by State
median_income_state_df = pd.DataFrame()

grouped_reduced_income_per_capita_df_2 = reduced_income_per_capita_df_2.groupby('State')

for column in reduced_income_per_capita_df_2[['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']]:
    series_median=grouped_reduced_income_per_capita_df_2[column].median()
    series_median_df = pd.DataFrame(series_median)
    median_income_state_df= median_income_state_df.append(series_median)

median_income_state_df.reset_index(inplace = True)
median_income_state_df
```

```python
population_df = reduced_BEA_df.loc[reduced_BEA_df['LineCode'] == 2.0]
population_df
```

```python
reduced_population_df = population_df.loc[(population_df['2013']!= "(NA)") & (population_df['2014']!= "(NA)")
                                                       & (population_df['2015']!= "(NA)") & (population_df['2016']!= "(NA)")
                                                       & (population_df['2017']!= "(NA)") & (population_df['2018']!= "(NA)") & (population_df['2019']!= "(NA)") & (population_df['2020']!= "(NA)")]

reduced_population_df.dropna(how = 'any', inplace = True)
reduced_population_df
```

```python
reduced_population_df[['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']] = reduced_population_df[['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']].apply(pd.to_numeric)
```

```python
# removing rows from the df with state name more than two characters

reduced_population_df['state_length'] = reduced_population_df.State.str.len()
reduced_population_df_2 = reduced_population_df.loc[reduced_population_df['state_length']<4]
reduced_population_df_2
```

```python
#Total population per State
state_population_df = pd.DataFrame()

grouped_reduced_population_df_2 = reduced_population_df_2.groupby('State')

for column in reduced_income_per_capita_df_2[['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']]:
    series_sum = grouped_reduced_population_df_2[column].sum()
    series_sum_df = pd.DataFrame(series_sum)
    state_population_df = state_population_df.append(series_sum)

state_population_df.reset_index(inplace = True)
state_population_df
```

```python
reduced_ht_df_fs = ht_df_fs[['Year', 'ACTUAL_COUNT', 'state_abb']]
reduced_ht_df_fs
```

```python
list_of_states_2 
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
```

```python
#Adding the states that are not included in the human traffiking dataframe to match it with BEA income and population data

for year in years:
    loc_df = reduced_ht_df_fs.loc[reduced_ht_df_fs['Year'] == year]
    for state in list_of_states_2 :
        
        if state not in loc_df['state_abb'].to_list():

            data = [{'Year':year,'ACTUAL_COUNT': 0, 'state_abb':state}]
            data_df = pd.DataFrame(data)
            reduced_ht_df_fs = reduced_ht_df_fs.append(data)
            
# reduced_ht_df_fs.sort_values(by=['Year', 'state_abb'], ascending = True)
reduced_ht_df_fs
```

```python
grouped_ht_df_fs = reduced_ht_df_fs.groupby(['Year', 'state_abb']).sum()['ACTUAL_COUNT']
grouped_ht_fs_df = pd.DataFrame(grouped_ht_df_fs)
grouped_ht_fs_df.reset_index(inplace = True)
grouped_ht_fs_df
```

```python
grouped_ht_fs_df_trans= grouped_ht_fs_df.pivot(index='Year', columns='state_abb', values='ACTUAL_COUNT')
grouped_ht_fs_df_trans
```

```python
grouped_ht_fs_df_trans.replace(np.nan,0)
grouped_ht_fs_df_trans.reset_index(inplace= True)
#grouped_ht_fs_df_trans.drop('index', axis=1, inplace=True)
grouped_ht_fs_df_trans
```
