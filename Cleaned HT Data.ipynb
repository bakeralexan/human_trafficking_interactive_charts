```python
csv_file = "Resources/HT_2013-2020.csv"
human_traffic_df = pd.read_csv(csv_file)
human_traffic_df.head()
```

```python
index_names = human_traffic_df[ human_traffic_df['AGENCY_TYPE_NAME'] == 'City' ].index

human_traffic_df.drop(index_names, inplace = True)
  
human_traffic_df
                       
```
