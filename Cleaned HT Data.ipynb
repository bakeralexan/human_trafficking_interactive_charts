```python
import pandas as pd
import numpy as np
import os
```

```python
csv_file = "Resources/HT_2013-2020.csv"
human_traffic_df = pd.read_csv(csv_file)
human_traffic_df.head()
```

```python
index_names = human_traffic_df[ human_traffic_df['AGENCY_TYPE_NAME'] == 'City' ].index

human_traffic_df.drop(index_names, inplace = True)
  
human_traffic_df
                       
```

```python
reduced_hf_df = human_traffic_df[["DATA_YEAR", "ORI", "STATE_NAME", "COUNTY_NAME",
                                    "REGION_NAME", "POPULATION_GROUP_CODE", "OFFENSE_NAME", "OFFENSE_SUBCAT_NAME", "ACTUAL_COUNT", "UNFOUNDED_COUNT", "CLEARED_COUNT", "JUVENILE_CLEARED_COUNT"]]
reduced_hf_df
```

```python
rename_ht_df = reduced_hf_df.rename(columns={'DATA_YEAR': 'Year', 'STATE_NAME': 'State', 'COUNTY_NAME': 'County', 'REGION_NAME': 'Region', 'POPULATION_GROUP_CODE': 'Population', 'OFFENSE_NAME': 'Offense Name', 'OFFENSE_SUBCAT_NAME': 'Offense Subcategory'})
rename_ht_df
```

```python
rename_ht_df.dropna(subset = ["County"], inplace=True)
rename_ht_df
```

```python
new_ht_df = rename_ht_df.loc[::, ['Year', 'County', 'State', 'Offense Subcategory', 'ACTUAL_COUNT']]
new_ht_df
```

```python
new_ht_df["County/State"] = new_ht_df["County"] + ','+ ' ' + new_ht_df["State"]
new_ht_df
```

```python
ht_2013 = new_ht_df.loc[new_ht_df['Year']== 2013]
ht_2013.head(40)
```

```python
sum_2013 = ht_2013.groupby('Year').sum()

sum_2013.head(40)
```
